## GOAL: 
Scrape link information from 4 main div sections on wikipedia
- From Today's featured article
- in the news
- Did you know
- on this day <br>
URL : https://en.wikipedia.org/wiki/Main_Page

Store data in dictionary: <br>
{'div identifier':[{title:'','text':'','href':''},],

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome()
driver.get('https://en.wikipedia.org/wiki/Main_Page')
```

```python
TFA_CSS_LOC = 'div#mp-tfa'
ITN_CSS_LOC = 'div#mp-itn'
DYK_CSS_LOC = 'div#mp-dyk'
OTD_CSS_LOC = 'div#mp-otd'

loc_list = [TFA_CSS_LOC,ITN_CSS_LOC,DYK_CSS_LOC,OTD_CSS_LOC]

link_dict = {}
```

```python
#iterate through CSS locators that identify each desired wiki section
for loc in loc_list:
    #find the div for the locator
    div_element = driver.find_element(By.CSS_SELECTOR,
                                     loc)
    info_lst = []
    #iterate through all the anchor elements in the selected div
    for link in div_element.find_elements(By.TAG_NAME,'a'):
        #capture title, text, href for each anchor
        info_dict = {
            'title':link.get_attribute('title'),
            'text':link.text,
            'href':link.get_attribute('href')
        }
        info_lst.append(info_dict)
    #store list of hrefs with data in dict
    link_dict[loc] = info_lst
```
